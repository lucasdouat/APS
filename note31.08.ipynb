Las señales causales son aquellas cuyo valor es nulo para tiempo anterior a un instante especifico, usualmente t=0. Es decir, un señal x(t) es causal si:
* En tiempo continuo: x(t)=0, para t<0
* En tiempo discreto x[n]=0 para todo n<0

  Esto significa que la señal no tiene "memoria del pasado". Antes del instante cero, lo cual es una propiedad importante en sistemas físicos y en procesamiento digital de señales, ya que muchos sistemas reales solo responden a estimulos presentes o futuros.

¿Que son las señales deterministicas?
Las señales deterministicas son aquellas cuyo comportamiento está completamente definido y puede predecirse con exactitud en cualquier instante de tiempo. Es decir, no hay incertidumbre sobre su evolución. si nocés la expresion matematica o la refla que la genera, podés calcular su valor en cualquier momento.

Contraste, señales aleatorias o estocásticas. Las señales aleatorias o estocásticas: tienen componentes impredecibles. No se puede conocer su valor exacto en un intante
